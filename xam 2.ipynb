```python
def seidal (a,x,b):
    # finding length of a(3)
    n = len(a)
    #for loop fro 3 times as to calculate x,y,z
    for j in range (0,n):
        # temp variable d to store b[j]
        d = b[j]
        
        # to calculate respective xi,yi,zi
        for i in range (0,n):
            if(j != i):
                d-=a[j][i] * x[i]
                # updtaing the value of our solution
        x[j] = d/ a[j][j]
    #returning our updated solution
    return x
G = np.zeros((11,3))
# input as number of variable to be solved
for k in range(0,1):
    n = 3 
    a = []
    b = []
    # initial solution depending on n (here n=3)
    x = [0,0,0]
    a = [[3,2,1],[2,3,-1],[1,-2,4]]
    b = [13+k,0,17]
  
    #loop for m times depending on m the error value
    for i in range (400,200):
        seidal( a,x,b)
    # print each time the updated solution
    print(x)
    G[k]=x
```
